```python
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
```

```python
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-1b7", use_cache=True)
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-1b7")
```

```python
set_seed(424242)
```

```python
 model.__class__.__name__
```

```python
def generate_prediction(prompt):
    # Tokenize the prompt
    input_ids = tokenizer(prompt, return_tensors="pt").to(0)

    # Get model prediction
    sample = model.generate(**input_ids, max_length=2, temperature=0, num_return_sequences=1)

    # Decode the generated sequence
    generated_sequence = tokenizer.decode(sample[0])

    return generated_sequence.strip()
```

```python
context = input("Enter the context: ")
question = input("Enter the question: ")
answer = input("Enter the answer: ")

# prompt = f"Context: {context} Question: {question} Answer: {answer}"
prompt = f"Context: {context}\nQuestion: {question}\nAnswer: {answer}\nIs this answer correct? (Answer with 'Yes' or 'No'): "
generate_prediction(prompt)
```
